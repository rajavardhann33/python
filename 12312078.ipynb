```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_csv("air quality and predition.csv")
df['Sampling Date'] = pd.to_datetime(df['Sampling Date'], errors='coerce')
df.dropna(subset=['Sampling Date'], inplace=True)
for col in ['SO2', 'NO2', 'SPM']:
    df[col] = df[col].fillna(df[col].mean())
```

```python
# 1 Scatter plot: NO2 vs PM10
plt.figure(figsize=(10, 6))
sns.s  catterplot(data=df, x='NO2', y='RSPM/PM10', hue='Type of Location', alpha=0.7,color="green")
plt.title('Scatter Plot: NO vs PM10 (RSPM)')
plt.xlabel('NO (g/m)')
plt.ylabel('PM10 (g/m)')
plt.grid(True)
plt.tight_layout()
```

```python
#2  Plotting the bar graph
df['Sampling Date'] = pd.to_datetime(df['Sampling Date'], errors='coerce')
df.dropna(subset=['Sampling Date'], inplace=True)
for col in ['SO2', 'NO2', 'SPM']:
    df[col] = df[col].fillna(df[col].mean())
avg_pollutants = df[['SO2', 'NO2', 'RSPM/PM10', 'SPM']].mean()
plt.figure(figsize=(8, 6))
avg_pollutants.plot(kind='bar', color='green', edgecolor='black')
plt.title('Average Levels of Pollutants')
plt.ylabel('Concentration (g/m)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# 3  Plot histograms
df['Sampling Date'] = pd.to_datetime(df['Sampling Date'], errors='coerce')
df.dropna(subset=['Sampling Date'], inplace=True)
for col in ['SO2', 'NO2', 'SPM']:
    df[col] = df[col].fillna(df[col].mean())
pollutants = ['SO2', 'NO2', 'RSPM/PM10', 'SPM']
plt.figure(figsize=(14, 10))
for i, pollutant in enumerate(pollutants, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[pollutant], bins=30, kde=True, color='cyan')
    plt.title(f'Distribution of {pollutant}')
    plt.xlabel(f'{pollutant} (g/m)')
    plt.ylabel('Frequency')
    plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#4 line plot date and calculate average NO2
df['Sampling Date'] = pd.to_datetime(df['Sampling Date'], errors='coerce')
df.dropna(subset=['Sampling Date'], inplace=True)
df['NO2'] = df['NO2'].fillna(df['NO2'].mean())
no2_trend = df.groupby('Sampling Date')['NO2'].mean()
plt.figure(figsize=(12, 6))
plt.plot(no2_trend.index, no2_trend.values, color='green', linewidth=2)
plt.title('NO Levels Over Time')
plt.xlabel('Date')
plt.ylabel('NO Concentration (g/m)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#5 the boxplot
df['Sampling Date'] = pd.to_datetime(df['Sampling Date'], errors='coerce')
df.dropna(subset=['Sampling Date'], inplace=True)
for col in ['SO2', 'NO2', 'SPM']:
    df[col] = df[col].fillna(df[col].mean())
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['SO2', 'NO2', 'RSPM/PM10', 'SPM']], palette="pastel")
plt.title('Box Plot of Pollutant Levels')
plt.ylabel('Concentration (g/m)')
plt.grid(True)
plt.tight_layout()
plt.show()
#6 Plot he
```

```python
#6 Plot heatmap
df['Sampling Date'] = pd.to_datetime(df['Sampling Date'], errors='coerce')
df.dropna(subset=['Sampling Date'], inplace=True)
for col in ['SO2', 'NO2', 'SPM']:
    df[col] = df[col].fillna(df[col].mean())
corr_matrix = df[['SO2', 'NO2', 'RSPM/PM10', 'SPM']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5,color="green")
plt.title('Correlation Heatmap of Pollutants')
plt.tight_layout()
plt.show()
```
